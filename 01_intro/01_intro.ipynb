````python
# data spliting
train_ind = np.random.choice(np.arange(len(clean_Article)),size = int(len(clean_Article)*0.7),replace = False)
train = clean_Article.iloc[train_ind].reset_index(drop = True)
test = clean_Article.iloc[~train_ind].reset_index(drop = True)

# create a set of word embedding. this makes each word as a columns
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(ngram_range = (1,2),max_features = 200).fit(train['concated_word_sentence_list'])
train = pd.concat([train,pd.DataFrame(vectorizer.transform(train['concated_word_sentence_list']).todense()).add_prefix('word_')],axis='columns')
test = pd.concat([test,pd.DataFrame(vectorizer.transform(test['concated_word_sentence_list']).todense()).add_prefix('word_')],axis='columns')
# extract some features from post date
today = pd.to_datetime('2019-11-11')
for df in [train,test]:
    df['days_since_post'] = (today - df['PostTime']).dt.days
    df['post_weekday'] =  df['PostTime'].dt.weekday
    df['post_month'] = df['PostTime'].dt.month
    df['post_season'] = df['PostTime'].dt.quarter
drop_cols = ['ArticleID','Subject','Thumbnail','PostTime','ReadBook','CreateUserID','CreateUserName','ModifyUserName','CreateTime','ModifyTime','concated_word_sentence_list','SubCategoryID']
processed_train = train.drop(drop_cols,axis = 'columns').fillna(0)
processed_test = test.drop(drop_cols,axis = 'columns').fillna(0)
#processed data
processed_X_train = processed_train.drop('ClickCount',axis = 'columns')
y_train = processed_train['ClickCount']
processed_X_test = processed_test.drop('ClickCount',axis = 'columns')
y_test = processed_test['ClickCount']
````

````python
processed_X_train.head()
````

### Linear Regression Algorithm

````python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(processed_X_train,y_train)
````

````python
# How good was the model fitted?
print("MAPE: %s" % np.mean(np.abs(y_train - lr_model.predict(processed_X_train))/y_train))
print("MAE: %s" % np.mean(np.abs(y_train - lr_model.predict(processed_X_train))))
````

````python
# How good was the model on predicting new click count for new articles?
print("MAPE: %s" % np.mean(np.abs(y_test - lr_model.predict(processed_X_test))/y_test))
print("MAE: %s" % np.mean(np.abs(y_test - lr_model.predict(processed_X_test))))
````

### Neural Network Algorithm

````python
from keras import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.losses import mean_absolute_error

nn_model = Sequential()
nn_model.add(Dense(units=64, input_dim=processed_X_train.shape[1],activation = 'relu')) 
nn_model.add(Dense(units=32)) 
nn_model.add(Dense(units=16))      
nn_model.add(Dense(units=1))
nn_model.compile(loss=mean_absolute_error, optimizer=Adam(lr=3e-3))
````

````python
# fit the NN model. this would take a while...
nn_model.fit(x=processed_X_train, y=y_train, batch_size=200, epochs=3000, verbose=0,validation_data = (processed_X_test,y_test))
````

````python
# How good was the model fitted on training data?
print("MAPE: %s" % np.mean(np.abs(y_train - nn_model.predict(processed_X_train)[:,0])/y_train))
print("MAE: %s" % np.mean(np.abs(y_train - nn_model.predict(processed_X_train)[:,0])))
````

````python
# How good was the model on predicting new click count for new articles?
print("MAPE: %s" % np.mean(np.abs(y_test - nn_model.predict(processed_X_test)[:,0])/y_test))
print("MAE: %s" % np.mean(np.abs(y_test - nn_model.predict(processed_X_test)[:,0])))
````

**Let's visualize the "learning process".** We could tell (1) the more the learning, the better the performance. (2) training error always smaller than testing error

````python
plt.plot(nn_model.history.history['loss'])
plt.plot(nn_model.history.history['val_loss'])
plt.title('MAE on log y')
plt.ylabel('MAE')
plt.xlabel('Epoch (# times the model learn on the data)')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
````

### RandomForest model

````python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators = 1000,n_jobs = -1)
rf_model.fit(processed_X_train,y_train)
print("MAPE: %s" % np.mean(np.abs(y_test - rf_model.predict(processed_X_test))/y_test))
print("MAE: %s" % np.mean(np.abs(y_test - rf_model.predict(processed_X_test))))
````

**Let's compare the train/test errrors in different complexity of RF models**

````python
# will take a moment
train_MAPE_list = []
train_MAE_list = []
test_MAPE_list = []
test_MAE_list = []
for i in range(3,72,3):
    model = RandomForestRegressor(n_estimators = 1000,max_depth = i,n_jobs = -1).fit(processed_X_train,y_train)
    
    train_MAPE_list.append(np.mean(np.abs(y_train - model.predict(processed_X_train))/y_train))
    train_MAE_list.append(np.mean(np.abs(y_train - model.predict(processed_X_train))))
    
    test_MAPE_list.append(np.mean(np.abs(y_test - model.predict(processed_X_test))/y_test))
    test_MAE_list.append(np.mean(np.abs(y_test - model.predict(processed_X_test))))
````

**Let's visualize the relation between model complexity and performance.** We could tell the more complex the model, the better the performance.

````python
plt.plot(train_MAPE_list)
plt.plot(test_MAPE_list)
plt.title('MAPE')
plt.ylabel('MAPE')
plt.xlabel('max_depth (model complexity)')
plt.legend(['Train', 'Test'], loc='upper left')
````

````python
plt.plot(train_MAE_list)
plt.plot(test_MAE_list)
plt.title('MAE')
plt.ylabel('MAE')
plt.xlabel('max_depth (model complexity)')
plt.legend(['Train', 'Test'], loc='upper left')
````
